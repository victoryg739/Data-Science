# Problem 1: Data Preparation

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
```

```python
train = pd.read_csv("train.csv",index_col="Id")
```

```python
train.head()
```

# What are the data types (“dtypes”) – int64/float64/object – of the variables (columns) in the dataset?

```python
train.dtypes
```

# Create a new Pandas DataFrame consisting of only the variables (columns) of type Integer (int64).

```python
train = train.select_dtypes(include='int64')
train.head()
```

# Open the “data_description.txt” file you downloaded (either from NTU Learn or Kaggle) in Wordpad.
# Read the description for each variable carefully and try to identify the “actual” Numeric variables.
# Categorical variables are often “encoded” as Numeric variables for easy representation. Spot them.

We identify following categorical variables that is represented as numeric variables
1. MSSubClass
2. YearBuilt (Does saying "The year 2000 is four times as much as the year 500" really make sense?)
3. YearRemodAdd
4. MoSold
5. YrSold
6. OverallQual
7. OverallCond

# Drop non-Numeric variables from the DataFrame to have a clean DataFrame with Numeric variables.

```python
train = train.drop(columns=['MSSubClass','YearBuilt','YearRemodAdd','MoSold','YrSold','OverallQual','OverallCond'])
train.head()
```

# Plot SalePrice (y-axis) vs LotArea (x-axis) using jointplot and find the Correlation between the two

```python
jointDF = pd.concat([train['SalePrice'], train['LotArea']], axis = 1).reindex(train['SalePrice'].index)

jointDF
```

```python
sb.jointplot(data = jointDF, x = "SalePrice", y = "LotArea", height = 12)
```
