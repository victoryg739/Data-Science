overallQual
Neighbourhood


```python
prob2Df = train[['MSSubClass','Neighborhood','BldgType','OverallQual','SalePrice']].copy()

fig, ax = plt.subplots(4,1, figsize=(20,40))
for var, subplot in zip(prob2Df, ax.flatten()):
    sorted_nb = train.groupby([var])['SalePrice'].median().sort_values()
    sb.boxplot(x=var, y='SalePrice', data=train, ax=subplot, order=list(sorted_nb.index))


```
